```python
# 문제1. 상품 재고 관리(딕셔너리 활용)

# 초기 재고 설정 (딕셔너리)
inventory = {
    "사과": 10,
    "바나나": 5,
    "체리": 8
}

print("초기 재고 상태")
print(inventory)

# 새로운 상품 추가 함수
def add_product(name, quantity):
    if name in inventory:
        print(f"오류 : '{name}'은 이미 재고에 있습니다.")
    else:
        inventory[name] = quantity
        print(f"'{name}' {quantity}개 추가 완료.")

# 특정 상품 재고 변경 함수
def update_stock(name, quantity):
    if name in inventory:
        inventory[name] = quantity
        print(f"'{name}'의 재고를 {quantity}개로 변경 완료.")
    else:
        print(f"오류 : '{name}'을(를) 찾을 수 없습니다.")

# 특정 상품 삭제 함수
def delete_product(name):
    if name in inventory:
        del inventory[name]
        print(f"'{name}' 삭제 완료.")
    else:
        print(f"오류 : '{name}'을(를) 찾을 수 없습니다.")

# 전체 상품 목록 조회 함수
def view_all_products():
    print("\n전체 재고 목록")
    if not inventory:
        print("재고가 비어있습니다.")
    else:
        for name, quantity in inventory.items():
            print(f"{name} : {quantity}개")

# 특정 상품 검색 함수
def search_product(name):
    print(f"\n'{name}' 검색 결과")
    if name in inventory:
        print(f"'{name}'의 재고는 {inventory[name]}개 입니다.")
    else:
        print(f"'{name}'은 재고에 없습니다.")

# 기능 실행 예시
print("\n기능 실행")
add_product("오렌지", 7)   # 새로운 상품 추가
update_stock("사과", 15)  # 재고 변경
delete_product("바나나")   # 상품 삭제
add_product("사과", 20)   # 이미 있는 상품 추가 시도
update_stock("포도", 10)  # 없는 상품 재고 변경 시도
delete_product("수박")    # 없는 상품 삭제 시도

# 최종 결과
view_all_products() # 전체 목록 조회
search_product("체리")    # 특정 상품 검색
```

```python
# 문제3. 도서 관리 시스템 (상속 및 다형성 활용)

# 부모 클래스: Book
class Book:
    # 생성자 : 책의 기본 속성인 제목과 저자를 초기화
    def __init__(self, title, author):
        self.title = title
        self.author = author

    # 정보 출력 메서드 (자식 클래스에서 재정의될 예정)
    def display_info(self):
        print(f"\n기본 책 정보")
        print(f"제목 : {self.title}")
        print(f"저자 : {self.author}")

# 자식 클래스 1: EBook (Book 클래스를 상속)
class EBook(Book):
    # 생성자 : 부모의 속성과 함께 전자책 고유 속성인 '파일 크기'를 초기화
    def __init__(self, title, author, file_size):
        # super()를 사용해 부모 클래스의 __init__ 메서드를 호출
        super().__init__(title, author)
        self.file_size = file_size

    # 메서드 오버라이딩(재정의) : 전자책에 맞는 정보 출력 방식으로 변경
    def display_info(self):
        print(f"\n전자책 정보")
        print(f"제목 : {self.title}")
        print(f"저자 : {self.author}")
        print(f"파일 크기 : {self.file_size}MB")

# 자식 클래스 2 : PaperBook (Book 클래스를 상속)
class PaperBook(Book):
    # 생성자 : 부모의 속성과 함께 종이책 고유 속성인 '페이지 수'를 초기화
    def __init__(self, title, author, pages):
        super().__init__(title, author)
        self.pages = pages

    # 메서드 오버라이딩(재정의) : 종이책에 맞는 정보 출력 방식으로 변경
    def display_info(self):
        print(f"\n종이책 정보")
        print(f"제목 : {self.title}")
        print(f"저자 : {self.author}")
        print(f"페이지 수 : {self.pages}쪽")

# 객체 생성 및 다형성 확인

# 각 클래스의 인스턴스(객체)를 생성
book_list = [
    EBook("EBook 테스트", "테스터", 50),
    PaperBook("PaperBook 테스트", "테스터2", 350),
    Book("Book 테스트", "테스터3")  # 부모 클래스 객체
]

# 반복문을 통해 각 객체의 display_info() 메서드를 호출
# 같은 이름의 메서드를 호출했지만, 각 객체의 실제 클래스에 따라 다른 결과가 출력. (다형성)
for book in book_list:
    book.display_info()
```

```python
# 문제 4. 주식 데이터(삼성전자, 카카오, 네이버) 를 활용하여 주가 변동을 시각화.

import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 한글 폰트 설정 (Windows)
font_path = "C:/Windows/Fonts/malgun.ttf"   # '맑은 고딕' 폰트 경로
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font_prop.get_name()  #Matplotlib 기본 폰트 설정
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호 깨짐 방지

#  종목 코드 설정
tickers = {
    '삼성전자': '005930.KS',
    '네이버': '035420.KS',
    '카카오': '035720.KS'
}

# 헤더 (야후 차단 방지용)
HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/124.0.0.0 Safari/537.36"
    )
}

try:
    # yfinance 라이브러리를 통해 종가(Close) 데이터만 가져옵니다.
    df_close = yf.download(list(tickers.values()), start='2022-01-01')['Close']
    df_close.columns = tickers.keys()   # 컬럼 이름을 한글 회사명으로 변경합니다.
    df_close = df_close.dropna()    # 데이터가 없는 날짜(결측치)는 제거합니다.
    print(df_close.head())

except Exception as e:
    print(f"데이터 수집 중 오류 발생: {e}")
    df_close = pd.DataFrame() # 오류 시 비어있는 상태로 초기화


# 데이터가 정상적으로 수집된 경우에만 아래의 시각화 코드를 실행합니다.
if not df_close.empty:
    # 2. 주가 변동 시각화 (라인 그래프)
    # 각 기업의 주가 변동을 라인 그래프로 시각화
    plt.figure(figsize=(14, 7))
    plt.plot(df_close.index, df_close['삼성전자'], label='삼성전자')
    plt.plot(df_close.index, df_close['네이버'], label='네이버')
    plt.plot(df_close.index, df_close['카카오'], label='카카오')

    plt.title('주요 IT 기업 주가 변동 추이')
    plt.xlabel('날짜')
    plt.ylabel('종가 (원)')
    plt.legend()
    plt.grid(True)
    plt.show()


# 이동 평균선 추가 및 시각화
# 주가의 단기(20일) 및 장기(60일) 추세를 파악하기 위해 이동 평균선을 계산하고 시각화
if not df_close.empty:
    plt.figure(figsize=(14, 7))

    # 원본 종가 그래프
    plt.plot(df_close.index, df_close['삼성전자'], label='삼성전자 종가', alpha=0.4)

    # 20일 이동 평균선 계산 및 그리기
    ma20 = df_close['삼성전자'].rolling(window=20).mean()
    plt.plot(ma20.index, ma20, label='20일 이동평균선', linestyle='--')

    # 60일 이동 평균선 계산 및 그리기
    ma60 = df_close['삼성전자'].rolling(window=60).mean()
    plt.plot(ma60.index, ma60, label='60일 이동평균선', linestyle=':')

    plt.title('삼성전자 주가와 이동 평균선')
    plt.xlabel('날짜')
    plt.ylabel('종가 (원)')
    plt.legend()
    plt.grid(True)
    plt.show()

# 상관관계 히트맵 시각화
# 각 종목 간 주가의 상관관계를 히트맵으로 시각화

if not df_close.empty:
    # 상관계수 행렬 계산
    corr_matrix = df_close.corr()

    plt.figure(figsize=(8, 6))
    # 히트맵 그리기
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)

    plt.title('기업별 주가 상관관계 히트맵')
    plt.show()
```
