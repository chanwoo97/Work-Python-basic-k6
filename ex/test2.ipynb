```python
from re import search

# 기본실습1 : 리스트 활용 (추가, 삭제, 수정, 조회, 검색)
# 문제 : 리스트를 활용하여 다음 기능을 수행하는 프로그램을 작성하세요.

# 빈 리스트를 하나 생성
my_list = []
print(f"초기상태 : {my_list}\n")

# 1. 추가 (append)
print("1. 추가")
my_list.append("사과")
my_list.append("바나나")
my_list.append("체리")
print(f"요소 추가 후 : {my_list}\n")

# 2. 수정 (인덱스 사용)
print("2. 수정")
# 0번 인덱스('사과')의 값을 '오렌지'로 변경
my_list[0] = "오렌지"
print(f"0번 인덱스 수정 후 : {my_list}\n")

# 3. 삭제 (remove)
print("3. 삭제")
# '바나나'라는 값을 찾아서 삭제
my_list.remove("바나나")
print(f"'바나나' 삭제 후 : {my_list}\n")

# 4. 조회 (전체 출력)
print("4. 조회")
print(f"현재 리스트의 모든 요소 : {my_list}\n")

# 5. 검색 (in 연산자)
print("5. 검색")
search_item = "체리"
if search_item in my_list:
    print(f"'{search_item}'는 리스트에 존재합니다.")
else:
    print(f"'{search_item}'는 리스트에 없습니다.")

search_item = "사과"
if search_item in my_list:
    print(f"'{search_item}'는 리스트에 존재합니다.")
else:
    print(f"'{search_item}'는 리스트에 없습니다.")
```

```python
# 기본실습2 : 튜플 활용 (조회, 검색)
# 문제 : 튜플을 활용하여 다음 기능을 수행하는 프로그램을 작성하세요.
# 주어진 튜플에서 특정 인덱스의 값을 조회
# 특정 값이 튜플에 있는지 검색
# 단순 기능 구현만 보이기

# 과일 이름이 담긴 튜플을 생성합니다.
my_tuple = ("사과", "바나나", "체리", "오렌지")
print(f"생성된 튜플 : {my_tuple}\n")

# 1. 조회 (인덱스 사용)
print("1. 조회")
# 2번 인덱스에 있는 값을 조회하여 출력
index_to_access = 2
value = my_tuple[index_to_access]
print(f"{index_to_access}번 인덱스의 값은 '{value}' 입니다.\n")

# 2. 검색 (in 연산자)
print("2. 검색")
search_item = "바나나"
if search_item in my_tuple:
    print(f"'{search_item}'는 튜플에 존재합니다.")
else:
    print(f"'{search_item}'는 튜플에 없습니다.")
```

```python
# 기본실습3 : 셋 활용 (추가, 삭제, 조회, 검색)
# 문제 : 셋을 활용하여 다음 기능을 수행하는 프로그램을 작성하세요.
# 사용자가 입력한 값을 셋에 추가
# 특정 값을 삭제
# 셋의 모든 요소 조회
# 특정 값이 셋에 있는지 검색
# 단순 기능 구현만 보이기

# 빈 셋을 하나 생성
my_set = set()
print(f"초기 상태 : {my_set}\n")

# 1. 추가 (add)
print("1. 추가")
my_set.add("사과")
my_set.add("바나나")
my_set.add("체리")
print(f"요소 추가 후 : {my_set}")

# 셋의 특징: 중복을 허용하지 않음
print("\n'사과'를 한 번 더 추가해 보겠습니다...")
my_set.add("사과")
print(f"중복된 값 추가 시도 후 (변화 없음) : {my_set}\n")

# 2. 삭제 (remove)
print("2. 삭제")
# '바나나'라는 값을 찾아서 삭제
my_set.remove("바나나")
print(f"'바나나' 삭제 후 : {my_set}\n")

# 3. 조회 (전체 출력)
# 셋은 순서가 없으므로, 출력 순서는 실행할 때마다 다를 수 있습니다.
print("3. 조회")
print(f"현재 셋의 모든 요소 : {my_set}\n")

# 4. 검색 (in 연산자)
# 셋은 내부적으로 해시 테이블을 사용하므로 검색 속도가 리스트보다 매우 빠릅니다.
print("4. 검색")
search_item = "체리"
if search_item in my_set:
    print(f"'{search_item}'는 셋에 존재합니다.")
else:
    print(f"'{search_item}'는 셋에 없습니다.")

search_item = "오렌지"
if search_item in my_set:
    print(f"'{search_item}'는 셋에 존재합니다.")
else:
    print(f"'{search_item}'는 셋에 없습니다.")
```

```python
# 응용실습1 : 학생 목록 관리 (리스트 활용)
# 문제 : 학생 이름을 저장하는 리스트를 활용하여 다음 기능을 구현하세요.
# 학생 추가
# 학생 삭제
# 학생 목록 출력
# 특정 학생이 목록에 있는지 검색
# 단순 기능 구현만 보이기

# 1. 학생 추가 함수
def add_student(student_list, name):
    student_list.append(name)
    print(f"'{name}' 학생을 목록에 추가했습니다.")

# 2. 학생 삭제 함수
def remove_student(student_list, name):
    if name in student_list:
        student_list.remove(name)
        print(f"'{name}' 학생을 목록에서 삭제했습니다.")
    else:
        print(f"삭제하려는 '{name}' 학생이 목록에 없습니다.")

# 3. 학생 목록 출력 함수
def print_student_list(student_list):
    print("\n--- 현재 학생 목록 ---")
    if not student_list:
        print("(학생이 없습니다)")
    else:
        for student in student_list:
            print(f"- {student}")
    print("--------------------")

# 4. 특정 학생 검색 함수
def search_student(student_list, name):
    if name in student_list:
        print(f"'{name}' 학생은 목록에 존재합니다.")
    else:
        print(f"'{name}' 학생은 목록에 존재하지 않습니다.")


# --- 기능 실행 ---

# 초기 학생 리스트 생성
students = []
print_student_list(students)

# 학생 추가
add_student(students, "김민준")
add_student(students, "이서연")
add_student(students, "박도윤")
print_student_list(students)

# 학생 삭제
remove_student(students, "이서연")
print_student_list(students)

# 없는 학생 삭제 시도
remove_student(students, "최지우")
print_student_list(students)

# 학생 검색
search_student(students, "김민준")
search_student(students, "이서연")
```

```python
# 응용실습2 : 상품 재고 관리 (딕셔너리 활용)
# 문제 : 딕셔너리를 활용하여 상품 재고를 관리하는 프로그램을 작성하세요.
# 새로운 상품 추가
# 특정 상품의 재고 변경
# 특정 상품 삭제
# 전체 상품 목록 조회
# 특정 상품 검색
# 단순 기능 구현만 보이기

# {상품명: 재고} 형태의 빈 딕셔너리를 생성
inventory = {}
print(f"초기 재고 상태 : {inventory}\n")

# 1. 새로운 상품 추가
print("1. 새로운 상품 추가")
inventory["사과"] = 10
inventory["바나나"] = 5
inventory["오렌지"] = 8
print(f"상품 추가 후: {inventory}\n")

# 2. 특정 상품의 재고 변경
print("2. 재고 변경")
item_to_update = "바나나"
new_stock = 3
inventory[item_to_update] = new_stock
print(f"'{item_to_update}'의 재고를 {new_stock}개로 변경 후 : {inventory}\n")

# 3. 특정 상품 삭제
print("3. 상품 삭제")
item_to_delete = "오렌지"
if item_to_delete in inventory:
    del inventory[item_to_delete]
    print(f"'{item_to_delete}' 상품 삭제 후 : {inventory}\n")
else:
    print(f"삭제하려는 '{item_to_delete}' 상품이 없습니다.\n")

# 4. 전체 상품 목록 조회
print("4. 전체 상품 목록 조회")
print("현재 전체 재고 현황 :")
if not inventory:
    print("(재고가 없습니다)")
else:
    # .items()를 사용하면 키와 값을 동시에 가져올 수 있습니다.
    for item, stock in inventory.items():
        print(f"- {item}: {stock}개")
print() # 한 줄 띄우기

# 5. 특정 상품 검색
print("5. 특정 상품 검색")
item_to_search = "사과"
if item_to_search in inventory:
    stock_count = inventory[item_to_search]
    print(f"'{item_to_search}'는 재고 목록에 있으며, 현재 {stock_count}개 있습니다.")
else:
    print(f"'{item_to_search}'는 재고 목록에 없습니다.")

item_to_search = "포도"
if item_to_search in inventory:
    stock_count = inventory[item_to_search]
    print(f"'{item_to_search}'는 재고 목록에 있으며, 현재 {stock_count}개 있습니다.")
else:
    print(f"'{item_to_search}'는 재고 목록에 없습니다.")
```

```python
# 응용실습3 : 유일한 값 추출기 (셋 활용)
# 문제 : 리스트에서 중복된 값을 제거하고, 유일한 값들만 저장하는 프로그램을 작성하세요.
# 단순 기능 구현만 보이기

# 1. 중복된 값을 포함하는 원본 리스트를 준비합니다.
original_list = ["사과", "바나나", "체리", "사과", "오렌지", "바나나", "사과"]
print(f"원본 리스트 (중복 포함) : {original_list}")
print(f"원본 리스트의 길이 : {len(original_list)}\n")


# 2. 리스트를 셋으로 변환하여 유일한 값만 추출합니다.
unique_set = set(original_list)

print("리스트를 셋으로 변환")
print(f"유일한 값만 남은 셋 : {unique_set}")
print(f"유일한 값의 개수 : {len(unique_set)}\n")


# 3. (선택 사항) 유일한 값들로 이루어진 새로운 리스트 만들기
# 필요하다면, 중복이 제거된 셋을 다시 리스트로 변환할 수 있습니다.
unique_list = list(unique_set)
print("셋을 다시 리스트로 변환")
print(f"최종 결과 (중복 제거된 리스트) : {unique_list}")
```
